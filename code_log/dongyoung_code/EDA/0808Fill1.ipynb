```python
import pandas as pd

# Load the data from the uploaded CSV file
file_path = 'trim_train_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
data.info()
```

```python
# Extract columns that contain 'Fill1' and the target variable
fill1_columns = [col for col in data.columns if 'Fill1' in col]
target_column = ['target']

# Create a new dataframe with these columns
fill1_target_data = data[fill1_columns + target_column]
```

```python
fill1_target_data
# Remove columns with any missing values
fill1_target_data_clean = fill1_target_data.dropna(axis=1, how='any')

fill1_target_data_clean
```

```python
# List of columns to remove
columns_to_remove = [
    'Wip Line_Fill1', 'Process Desc._Fill1', 'Equipment_Fill1',
    'Insp. Seq No._Fill1', 'Insp Judge Code_Fill1', 
    'Production Qty Collect Result_Fill1', 'Receip No Collect Result_Fill1'
]

# Remove the specified columns
cleaned_data = fill1_target_data_clean.drop(columns=columns_to_remove)
```

```python
cleaned_data
```

```python
import matplotlib.pyplot as plt

# Group by 'Model.Suffix_Fill1' and 'target' and calculate the size
count_df = cleaned_data.groupby(['Model.Suffix_Fill1', 'target']).size().unstack(fill_value=0)

# Calculate the proportion for each 'Model.Suffix_Fill1'
count_df = count_df.div(count_df.sum(axis=1), axis=0)

# Plot the proportions
count_df.plot(kind='bar', stacked=True, color=['salmon', 'skyblue'], figsize=(8, 4))

plt.xlabel('Model Suffix')
plt.ylabel('Proportion')
plt.title('Proportion of Target by Model Suffix')
plt.legend(title='Target', loc='upper right')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add the proportion text on the bars
for i in range(count_df.shape[0]):
    for j in range(count_df.shape[1]):
        plt.text(i, count_df.iloc[i, j] / 2 + count_df.iloc[i, :j].sum(), f"{count_df.iloc[i, j]:.2f}", ha='center', color='white')

# Display the plot
plt.show()
```
